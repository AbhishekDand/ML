# Merge and Join

Joins can be one-to-one, one-to-many and many-to-many

```python
df6= pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})
```

```python
df7 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],
'hire_date': [2004, 2008, 2012, 2014]})
df7
```

```python
df6
```

```python
#one-to-one join
df8=pd.merge(df6,df7)
```

```python
df8
```

```python
#many-to-one join
df9 = pd.DataFrame({'group': ['Accounting', 'Engineering', 'HR'],
'supervisor': ['Carly', 'Guido', 'Steve']})
```

```python
df9
```

```python
print(pd.merge(df6,df9))
```

```python
#on keyword for having merge with respect to specific column
pd.merge(df6,df7,on='employee')
```

Getting similar table using pivot table

```python
data.pivot_table('BMI',index='Outcome',columns='Pregnancies')
```
